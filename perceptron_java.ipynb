```python
import java.util.Random;

public class Perceptron {

    private double[] weights;
    private double bias;
    private double learningRate;
    private Random random;
    private ActivationFunction activationFunction;

    public enum ActivationFunction {
        LINEAR, STEP, SIGMOID, RELU, SOFTMAX
    }

    public Perceptron(int inputSize, ActivationFunction activationFunction) {
        weights = new double[inputSize];
        random = new Random();
        this.activationFunction = activationFunction;

        // Initialize weights and bias randomly between -0.5 and 0.5
        for (int i = 0; i < inputSize; i++) {
            weights[i] = random.nextDouble() - 0.5;
        }
        bias = random.nextDouble() - 0.5;

        // Learning rate between 0 and 0.1
        learningRate = random.nextDouble() * 0.1;

        System.out.println("Initial weights and bias:");
        printWeights();
        System.out.printf("Learning rate (alpha): %.4f\n\n", learningRate);
    }

    public void train(double[][] data, int[] target, int epochs) {
        if (epochs <= 10) {
            throw new IllegalArgumentException("Number of epochs must be greater than 10");
        }

        for (int epoch = 0; epoch < epochs; epoch++) {
            int errors = 0;
            System.out.printf("===== Epoch %d =====\n", epoch + 1);

            for (int i = 0; i < data.length; i++) {
                int output = predict(data[i]);
                int error = target[i] - output;

                if (error != 0) {
                    for (int j = 0; j < weights.length; j++) {
                        weights[j] += learningRate * error * data[i][j];
                    }
                    bias += learningRate * error;
                    errors++;
                }

                System.out.printf("Pattern %d -> Expected: %d, Predicted: %d, Error: %d\n",
                        i + 1, target[i], output, error);
            }

            System.out.printf("Errors in epoch %d: %d\n\n", epoch + 1, errors);

            if (errors == 0) {
                System.out.println("Training completed without errors!\n");
                break;
            }
        }
    }

    public int predict(double[] inputs) {
        double sum = 0;
        for (int i = 0; i < weights.length; i++) {
            sum += inputs[i] * weights[i];
        }
        sum += bias;

        switch (activationFunction) {
            case LINEAR:
                return (int) linear(sum);
            case STEP:
                return step(sum);
            case SIGMOID:
                return sigmoid(sum) >= 0.5 ? 1 : 0;
            case RELU:
                return (int) relu(sum);
            case SOFTMAX:
                // Softmax requires multi-class output; we'll keep it simple for binary output.
                return softmax(new double[]{sum})[0] > 0.5 ? 1 : 0;
            default:
                throw new IllegalArgumentException("Invalid activation function");
        }
    }

    private double linear(double x) {
        return x;
    }

    private int step(double x) {
        return x >= 0 ? 1 : 0;
    }

    private double sigmoid(double x) {
        return 1.0 / (1.0 + Math.exp(-x));
    }

    private double relu(double x) {
        return Math.max(0, x);
    }

    private double[] softmax(double[] inputs) {
        double max = Double.NEGATIVE_INFINITY;
        for (double input : inputs) {
            if (input > max) max = input;
        }

        double sum = 0.0;
        double[] exp = new double[inputs.length];
        for (int i = 0; i < inputs.length; i++) {
            exp[i] = Math.exp(inputs[i] - max);
            sum += exp[i];
        }

        double[] result = new double[inputs.length];
        for (int i = 0; i < inputs.length; i++) {
            result[i] = exp[i] / sum;
        }
        return result;
    }

    public void printWeights() {
        for (int i = 0; i < weights.length; i++) {
            System.out.printf("w[%d] = %.4f\n", i + 1, weights[i]);
        }
        System.out.printf("Bias = %.4f\n\n", bias);
    }

    public static void main(String[] args) {
        double[][] data = {
            {20, 70, 1013, 15, 180, 0, 60},
            {25, 60, 1010, 10, 90, 0, 40},
            {30, 80, 1015, 20, 270, 5, 70},
            {15, 90, 1005, 5, 0, 10, 90},
            {10, 50, 1012, 12, 45, 0, 20}
        };

        int[] target = {1, 1, 0, 0, 1};

        // Choose the activation function here:
        Perceptron perceptron = new Perceptron(7, ActivationFunction.SOFTMAX);

        // Train with at least 11 iterations
        perceptron.train(data, target, 20);

        System.out.println("===== Test =====");
        for (int i = 0; i < data.length; i++) {
            int output = perceptron.predict(data[i]);
            System.out.printf("Input: (%.2f, %.2f, %.2f, %.2f, %.2f, %.2f, %.2f) -> Output: %d (Expected: %d)\n",
                    data[i][0], data[i][1], data[i][2], data[i][3], data[i][4], data[i][5], data[i][6],
                    output, target[i]);
        }
    }
}
```
