```python
from transformers import pipeline
import gradio as gr

# Cargar modelo preentrenado para análisis de sentimiento
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# Función que se conectará a la interfaz
def analizar_sentimiento(texto):
    resultado = classifier(texto)[0]
    etiqueta = resultado['label']
    puntuacion = resultado['score']
    return f"Sentimiento: {etiqueta} ({puntuacion:.2f})"

# Crear la interfaz web
iface = gr.Interface(
    fn=analizar_sentimiento,
    inputs=gr.Textbox(lines=3, placeholder="Escribe una opinión..."),
    outputs="text",
    title="Clasificador de Sentimientos",
    description="Ingresa una frase y obtén si es Positiva o Negativa usando un modelo Transformer"
)

# Ejecutar localmente
if __name__ == "__main__":
    iface.launch()
```
