```python
import random
import math

datos_entrenamiento = [
    ([30, 70, 1013, 10, 180, 2, 60], 1),  # Clima favorable
    ([25, 80, 1010, 5, 200, 5, 80], 0),   # Clima desfavorable
    ([35, 60, 1015, 15, 150, 0, 40], 1),  # Clima favorable
    ([20, 90, 1008, 2, 250, 10, 90], 0),  # Clima desfavorable
    ([32, 65, 1012, 12, 160, 1, 50], 1),  # Clima favorable
    ([28, 75, 1011, 8, 190, 3, 70], 0),   # Clima desfavorable
    ([36, 55, 1016, 18, 140, 0, 30], 1),  # Clima favorable
    ([22, 85, 1009, 4, 230, 7, 85], 0),   # Clima desfavorable
    ([31, 68, 1014, 11, 170, 2, 55], 1),  # Clima favorable
    ([26, 78, 1010, 6, 210, 4, 75], 0),   # Clima desfavorable
    ([38, 50, 1017, 20, 130, 0, 20], 1),  # Clima favorable
    ([19, 92, 1007, 1, 270, 12, 95], 0),  # Clima desfavorable
    ([34, 62, 1013, 14, 155, 1, 45], 1),  # Clima favorable
    ([27, 77, 1010, 7, 195, 3, 68], 0),   # Clima desfavorable
    ([37, 53, 1016, 19, 135, 0, 25], 1),  # Clima favorable
    ([21, 88, 1008, 3, 240, 9, 88], 0),   # Clima desfavorable
    ([33, 66, 1012, 13, 165, 2, 52], 1),  # Clima favorable
    ([29, 72, 1011, 9, 185, 3, 65], 0),   # Clima desfavorable
    ([39, 48, 1018, 22, 120, 0, 15], 1),  # Clima favorable
    ([18, 95, 1006, 0, 280, 15, 98], 0)   # Clima desfavorable
]

tasa_aprendizaje = 0.1
epochs = 100

# --------------------------
# Definición de funciones de activación y sus derivadas cuando corresponde

# Función lineal: f(z) = z
def linear(z):
    return z

def deriv_linear(z, output):
    return 1

# Función escalón: f(z) = 1 si z >= 0, 0 en caso contrario.
def escalon(z):
    return 1 if z >= 0 else 0
# (No se utiliza derivada pues la función escalón no es diferenciable; se usará regla clásica del perceptrón).

# Función sigmoidal: f(z) = 1/(1 + exp(-z))
def sigmoidal(z):
    return 1 / (1 + math.exp(-z))

def deriv_sigmoidal(z, output):
    # Derivada en función de la salida: f'(z) = output * (1 - output)
    return output * (1 - output)

# Función ReLU: f(z) = max(0, z)
def relu(z):
    return max(0, z)

def deriv_relu(z, output):
    return 1 if z > 0 else 0

# Función tangente hiperbólica: f(z) = tanh(z)
def tangente_hiperbolica(z):
    return math.tanh(z)

def deriv_tanh(z, output):
    return 1 - output**2

# Función softmax: Se implementa para un vector z, retornando una lista de probabilidades.
def softmax(z_list):
    max_z = max(z_list)  # Evitar overflow restando el máximo
    exps = [math.exp(z - max_z) for z in z_list]
    sum_exps = sum(exps)
    return [exp_val / sum_exps for exp_val in exps]

# --------------------------
# Entrenamiento para neuronas de una salida
# (Se entrena usando el mismo dataset, pero aplicando la función de activación indicada.)
def train_single_neuron(nombre_func, activation_func, deriv_func=None):
    print(f"\n--- Entrenamiento con función {nombre_func} ---")
    # Inicialización aleatoria de pesos y bias
    pesos = [random.uniform(-0.5, 0.5) for _ in range(7)]
    bias = random.uniform(-0.5, 0.5)

    for epoch in range(epochs):
        total_error = 0
        for entradas, salida_deseada in datos_entrenamiento:
            # Cálculo de la suma ponderada
            z = sum(entrada * peso for entrada, peso in zip(entradas, pesos)) + bias
            output = activation_func(z)
            error = salida_deseada - output
            total_error += abs(error)

            # Si se tiene derivada, se usa actualización basada en gradiente descendente
            if deriv_func is not None:
                # En funciones como sigmoidal, tanh y ReLU, usamos la derivada
                delta = error * deriv_func(z, output)
            else:
                # Para escalón se usa la regla clásica del perceptrón (no diferenciable)
                delta = error

            # Actualización de pesos y bias
            for i in range(len(pesos)):
                pesos[i] += tasa_aprendizaje * delta * entradas[i]
            bias += tasa_aprendizaje * delta

        print(f"Epoch {epoch+1}, Error Total: {total_error:.4f}, Pesos: {['{:.3f}'.format(w) for w in pesos]}, Bias: {bias:.3f}")
        if total_error == 0:
            print("Entrenamiento finalizado, modelo entrenado correctamente.")
            break

# --------------------------
# Entrenamiento para softmax con dos salidas
# Convertimos la salida deseada en vector one-hot: clase 0 -> [1, 0], clase 1 -> [0, 1].
def train_softmax():
    print("\n--- Entrenamiento con función Softmax ---")
    # Inicialización de pesos para dos neuronas
    pesos0 = [random.uniform(-0.5, 0.5) for _ in range(7)]
    pesos1 = [random.uniform(-0.5, 0.5) for _ in range(7)]
    bias0 = random.uniform(-0.5, 0.5)
    bias1 = random.uniform(-0.5, 0.5)

    for epoch in range(epochs):
        total_error = 0
        for entradas, salida_deseada in datos_entrenamiento:
            # Salidas deseadas en formato one-hot
            if salida_deseada == 0:
                target = [1, 0]
            else:
                target = [0, 1]

            # Calcular z para cada neurona
            z0 = sum(x * w for x, w in zip(entradas, pesos0)) + bias0
            z1 = sum(x * w for x, w in zip(entradas, pesos1)) + bias1

            outputs = softmax([z0, z1])
            # Error es la diferencia target - output para cada neurona
            error0 = target[0] - outputs[0]
            error1 = target[1] - outputs[1]
            total_error += abs(error0) + abs(error1)

            # Actualización de pesos y bias (derivada de la entropía cruzada con softmax resulta en error directo)
            for i in range(len(pesos0)):
                pesos0[i] += tasa_aprendizaje * error0 * entradas[i]
                pesos1[i] += tasa_aprendizaje * error1 * entradas[i]
            bias0 += tasa_aprendizaje * error0
            bias1 += tasa_aprendizaje * error1

        print(f"Epoch {epoch+1}, Error Total: {total_error:.4f}, Pesos0: {['{:.3f}'.format(w) for w in pesos0]}, Bias0: {bias0:.3f}")
        print(f"              , Pesos1: {['{:.3f}'.format(w) for w in pesos1]}, Bias1: {bias1:.3f}")
        if total_error == 0:
            print("Entrenamiento finalizado, modelo entrenado correctamente.")
            break

# --------------------------
# Ejecutar entrenamientos para cada función de activación

# 1. Función Lineal
train_single_neuron("Lineal", linear, deriv_linear)

# 2. Función Escalón
# Aquí no se utiliza derivada (se usa la regla del perceptrón clásico)
train_single_neuron("Escalón", escalon, deriv_func=None)

# 3. Función Sigmoidal
train_single_neuron("Sigmoidal", sigmoidal, deriv_sigmoidal)

# 4. Función ReLU
train_single_neuron("ReLU", relu, deriv_relu)

# 5. Función Tangente Hiperbólica
train_single_neuron("Tangente Hiperbólica", tangente_hiperbolica, deriv_tanh)

# 6. Función Softmax (utilizando dos salidas)
train_softmax()
```

```python
#include <stdio.h>
#include <stdlib.h>
#include <math.h>

#define N_SAMPLES 10
#define N_FEATURES 7
#define EPOCHS 50
#define LEARNING_RATE 0.1f

// 1. Función de activación Escalón (no requiere funciones externas)
int step_activation(float x) {
    return (x >= 0.0f) ? 1 : 0;
}

// 2. Función de activación Sigmoide usando <math.h>
int sigmoid_activation(float x) {
    float s = 1.0f / (1.0f + expf(-x));
    return (s >= 0.5f) ? 1 : 0;
}

// 3. Función de activación tanh usando <math.h>
int tanh_activation(float x) {
    return (tanhf(x) >= 0.0f) ? 1 : 0;
}

// 4. Función de activación Leaky ReLU sin usar <math.h>
//    La función ReLU es lineal (f(x) = x para x>=0 y f(x)=alpha*x para x<0),
//    por lo que su serie de Taylor es trivial; aquí se implementa manualmente.
int leaky_relu_activation(float x) {
    float alpha = 0.01f;  // Coeficiente pequeño para x negativos
    float y;
    if (x >= 0.0f)
        y = x;
    else
        y = alpha * x;
    return (y > 0.0f) ? 1 : 0;
}

// Tipo para puntero a función de activación
typedef int (*activation_func)(float);

// Función para obtener el valor absoluto de un entero
int my_abs(int x) {
    return (x < 0) ? -x : x;
}

// Función de entrenamiento con early stopping (se detiene si total_error es 0)
void train_with_error(activation_func act_func, float weights[], float *bias,
                      float X[][N_FEATURES], int y[], const char *name) {
    printf("Entrenamiento del Perceptrón (%s):\n", name);
    int epoch, i, j;
    for (epoch = 0; epoch < EPOCHS; epoch++) {
        int total_error = 0;
        for (i = 0; i < N_SAMPLES; i++) {
            float activation = *bias;
            for (j = 0; j < N_FEATURES; j++)
                activation += weights[j] * X[i][j];

            int pred = act_func(activation);
            int error = y[i] - pred;
            total_error += my_abs(error);

            // Actualización de pesos y sesgo
            for (j = 0; j < N_FEATURES; j++)
                weights[j] += LEARNING_RATE * error * X[i][j];
            *bias += LEARNING_RATE * error;
        }
        printf("Epoch %d: Error Total = %d\n", epoch + 1, total_error);
        if (total_error == 0) {
            printf("Convergencia alcanzada en epoch %d. Finalizando entrenamiento.\n", epoch + 1);
            break;
        }
    }
    printf("\n");
}

// Función de predicción usando la función de activación dada
int predict_with_activation(activation_func act_func, float weights[], float bias, float features[]) {
    float sum = bias;
    int i;
    for (i = 0; i < N_FEATURES; i++)
        sum += weights[i] * features[i];
    return act_func(sum);
}

int main() {
    // Dataset sintético: cada fila representa [Temperatura, HR, Presión, Velocidad, Dirección, Precipitación, Nubosidad]
    float X[N_SAMPLES][N_FEATURES] = {
        {20.5, 65, 1015, 3.2,  90, 0.0, 10},
        {22.0, 72, 1012, 5.1, 180, 0.5, 50},
        {18.0, 80, 1008, 2.8, 270, 1.2, 90},
        {25.3, 55, 1010, 4.0, 135, 0.0, 20},
        {30.1, 40, 1005, 6.7,  45, 0.0,  5},
        {16.8, 85, 1007, 1.5, 300, 2.0, 100},
        {28.4, 60, 1013, 3.9, 210, 0.0, 30},
        {21.0, 75, 1011, 4.5, 150, 0.8, 70},
        {19.5, 68, 1009, 2.2, 330, 0.1, 40},
        {23.2, 50, 1014, 5.0, 120, 0.0, 15}
    };

    // Etiquetas: 0 o 1 (definidas sintéticamente)
    int y[N_SAMPLES] = {0, 1, 1, 0, 0, 1, 0, 1, 0, 0};

    // Inicialización de pesos y sesgos para cada perceptrón
    float weights_step[N_FEATURES] = {0};
    float bias_step = 0;

    float weights_sigmoid[N_FEATURES] = {0};
    float bias_sigmoid = 0;

    float weights_tanh[N_FEATURES] = {0};
    float bias_tanh = 0;

    float weights_leaky_relu[N_FEATURES] = {0};
    float bias_leaky_relu = 0;

    // Entrenar cada perceptrón (máximo 50 épocas, early stopping si error total es 0)
    train_with_error(step_activation, weights_step, &bias_step, X, y, "Escalón");
    train_with_error(sigmoid_activation, weights_sigmoid, &bias_sigmoid, X, y, "Sigmoide");
    train_with_error(tanh_activation, weights_tanh, &bias_tanh, X, y, "tanh");
    train_with_error(leaky_relu_activation, weights_leaky_relu, &bias_leaky_relu, X, y, "Leaky ReLU");

    // Dato de prueba
    float test[N_FEATURES] = {24.0, 78, 1010, 3.5, 160, 0.7, 60};

    // Realizar predicciones con cada perceptrón
    printf("Predicción para el dato de prueba:\n");
    printf("Perceptrón Escalón:    %d\n", predict_with_activation(step_activation, weights_step, bias_step, test));
    printf("Perceptrón Sigmoide:   %d\n", predict_with_activation(sigmoid_activation, weights_sigmoid, bias_sigmoid, test));
    printf("Perceptrón tanh:       %d\n", predict_with_activation(tanh_activation, weights_tanh, bias_tanh, test));
    printf("Perceptrón Leaky ReLU: %d\n", predict_with_activation(leaky_relu_activation, weights_leaky_relu, bias_leaky_relu, test));

    return 0;
}
```
